```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
```

```python
def f (x):
    return 1/np.sqrt(1+np.e**(-(x)**2))
def df(x):
    return ((np.e**(-(x)**2)*x))/(1+np.e**(-(x)**2))**(3/2)
```

```python
xi, xf, h = -10.0, 10.0, 0.05
Npns = int((xf-xi)/h)

x = np.linspace(xi,xf,Npns)
```

```python
def CDf (f,x,h):
    return (f(x+h)-f(x-h))/(2*h)
      
    
```

```python
Dy=df(x)
CDy =CDf(f,x,h)
```

```python
Sum_num = 0.
Sum_den = 0.

Error_l = np.abs(CDy-Dy)

for i in range(len(Error_l)):
    
    Sum_num += Error_l[i]**2
    Sum_den += Dy[i]**2
    
ErrorGlobal = np.sqrt(Sum_num / Sum_den)
```
