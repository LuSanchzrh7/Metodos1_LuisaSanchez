```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
def l_cmonedas(n,k_m):
  pos=[1,-1]
  frec=0
  for i in range(n):
    sale= random.choices(pos,k=k_m)
    caras= sale.count(1)
    if caras==2:
      frec+=1
  prob=frec/i
  return prob
l_cmonedas(int(1e5),4),l_cmonedas(int(1e4),4),l_cmonedas(int(1e7),4) #esta cerca de 0.37
```

## Punto 9

```python
m1 = np.linspace(0.1,0.5,100)
m2 = np.linspace(0.1,0.9,100)
```

```python
P_1,P_2=np.meshgrid(m1,m2)
```

```python
def calcular_arbol(P1,P2): #revisar demostracion en el pdf enviado
  probabilidad_ccss=((P1+P2)/4-((P1*P2)/2)+1/4)
  return probabilidad_ccss
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(P_1, P_2,calcular_arbol(P_1,P_2) , rstride=1, cstride=1, cmap=plt.cm.coolwarm ,linewidth=0, antialiased=False)
```
